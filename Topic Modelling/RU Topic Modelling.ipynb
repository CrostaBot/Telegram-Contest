```python
import fasttext
import pandas as pd
```

```python
df = pd.read_csv('RU20000.csv')
df = df.drop_duplicates(subset=['Messages'])
df = df.drop(['Unnamed: 0'], axis=1)

df
```

```python
df_en = pd.read_csv('ENRU20000.csv')
df_en = df_en.replace('#VALUE!', ' ')

df_en
```

```python
class Dataset:
    def __init__(self, title, description, message, label):
        self.title = title
        self.description = description
        self.message = message
        self.label = label
```

```python
modelEN = fasttext.load_model('Supervised_EN.bin')
dataset = []

for i in range(0, len(df_en)):
    string = str(df_en.iloc[i, 0]) + ' ' + str(df_en.iloc[i, 1]) +' '+ str(df_en.iloc[i, 2])
    prediction = modelEN.predict(string)
    prediction = prediction[0][0].replace('__label__', '')
    labeled  = Dataset(df_en.iloc[i, 0], df_en.iloc[i, 1], df_en.iloc[i, 2], prediction)
    dataset.append(labeled)
    
print(len(dataset))
```

```python
lab = []
for item in dataset:
    lab.append(str(item.label))

s1 = pd.Series(lab, name="Label")

result = pd.concat([df, s1], axis=1)

print(result['Label'].value_counts())

result
```

```python
with open("Category_RU.txt", "w") as text_file:
    for i in range(0, len(result)):
        string = '__label__' + str(result.iloc[i, 3]) + ' ' + str(result.iloc[i, 0]) + ' ' + str(result.iloc[i, 1]) + ' ' + str(result.iloc[i, 2]) + '\n'
        text_file.write(string)

with open('Category_RU.txt') as f:
    count = sum(1 for _ in f)

print(count)
        
```

```python
#create .train and .val
! head -n 18000 Category_RU.txt > Category_RU.train
! tail -n 2000 Category_RU.txt > Category_RU.valid
```

```python
model_RU = fasttext.train_supervised(input="Category_RU.train", epoch=50, lr=1.0, wordNgrams=3)
model_RU.save_model("Supervised_RU.bin")
model_RU.test("Category_RU.valid")
```

```python
dataset = []

for i in range(0, len(df_final)):
    string = str(df_final.iloc[i, 0]) + ' ' + str(df_final.iloc[i, 1]) +' '+ str(df_final.iloc[i, 2])
    prediction = model_RU.predict(string)
    prediction = prediction[0][0].replace('__label__', '')
    labeled  = Dataset(df_final.iloc[i, 0], df_final.iloc[i, 1], df_final.iloc[i, 2], prediction)
    dataset.append(labeled)
    
print(len(dataset))

```

```python
#write in txt dataset  , CREATE MODEL AND TRAIN PREDICT IN RU DATASET GIANT AND RETRAIN NEW MODEL
tit, descr, msg, lab = [], [], [], [] 
for item in dataset:
    tit.append(str(item.title))
    descr.append(str(item.description))
    msg.append(str(item.message))
    lab.append(str(item.label))

dataframe = {'Title': tit,
            'Description': descr,
            'Messages': msg,
            'Label': lab
            }

df_ru_total = pd.DataFrame(dataframe, columns = ['Title', 'Description', 'Messages', 'Label'])

with open("Category_RU_Final.txt", "w") as text_file:
    for i in range(0, len(df_ru_total)):
        string = '__label__' + str(df_ru_total.iloc[i, 3]) + ' ' + str(df_ru_total.iloc[i, 0]) + ' ' + str(df_ru_total.iloc[i, 1]) + ' ' + str(df_ru_total.iloc[i, 2]) + '\n'
        text_file.write(string)

with open('Category_RU_Final.txt') as f:
    count = sum(1 for _ in f)

print(count)
        
```

```python
newmodel = fasttext.train_supervised(input="Category_RU_Final.train", epoch=100, lr=1.0, wordNgrams=2)
newmodel.save_model("Supervised_RU_Final.bin")
newmodel.test("Category_RU_Final.valid")
```
