```python
import fasttext
import pandas as pd
```

```python
df = pd.read_csv('/Users/Nicola/Contest/Language Recognition/EN_Data.csv')
df = df.drop_duplicates(subset=['Messages'])
df['Label'] = 'ciao'

df
```

```python
df_label = pd.read_csv('test2.csv', sep =';')
df_label = df_label.sample(frac=1)
df_label = df_label.drop(['Unnamed: 0'], axis=1)
df_label = df_label.dropna(subset=['Label'])
labels = df_label['Label'].unique()
print(labels)

df_label
```

```python
df_label = df_label.loc[df_label['Label'] != 'ciao']
print(len(df_label))

print(df_label['Label'].unique())
```

```python
for i in range(0, len(df)):
    for j in range(0, len(df_label)):
            if df_label.iloc[j, 0] == df.iloc[i, 0]:
                df.iloc[i, 3] = df_label.iloc[j, 3]

df
```

```python
df_1 = df[df.Label != 'ciao']

print(len(df_1))
print(df_1.loc[df_1['Label'] == 'ciao'])
print(df_1['Label'].unique())

df_1
```

```python
df_ru = pd.read_csv('RUTEST.csv', sep = ";")
df_ru = df_ru.drop(['Unnamed: 0'], axis=1)
df_ru = df_ru[df_ru.Label != 'ciao']
df_ru = df_ru.dropna(subset=['Label'])

df_ru
```

```python
frames = [df_1, df_ru]
result = pd.concat(frames)
result = result.sample(frac=1)

print(result['Label'].value_counts())
print(len(result['Label'].value_counts()))
print(len(result))

result
```

```python
with open("Category_EN.txt", "w") as text_file:
    for i in range(0, len(result)):
        string = '__label__' + str(result.iloc[i, 3]) + ' ' + str(result.iloc[i, 0]) + ' ' + str(result.iloc[i, 1]) + ' ' + str(result.iloc[i, 2]) + '\n'
        text_file.write(string)

with open('Category_EN.txt') as f:
    count = sum(1 for _ in f)

print(count)

#create .train and .val
print(int(count*0.85))
print(count - int(count*0.85))
```

```python
! head -n 1116 Category_EN.txt > Category_EN.train
! tail -n 197 Category_EN.txt > Category_EN.valid
```

```python
model = fasttext.train_supervised(input="Category_EN.train", epoch=200, lr=1.0, wordNgrams=2)
model.save_model("Semisupervised_EN.bin")
model.test("Category_EN.valid")
```

```python
model.get_labels()
```

```python
class Dataset:
    def __init__(self, title, description, message, label):
        self.title = title
        self.description = description
        self.message = message
        self.label = label
```

```python
df_2 = df[df.Label == 'ciao']

dataset = []

for i in range(0, len(df_2)):
    string = str(df_2.iloc[i, 0]) + ' ' + str(df_2.iloc[i, 1]) +' '+ str(df_2.iloc[i, 2])
    prediction = model.predict(string)
    prediction = prediction[0][0].replace('__label__', '')
    labeled  = Dataset(df_2.iloc[i, 0], df_2.iloc[i, 1], df_2.iloc[i, 2], prediction)
    dataset.append(labeled)
    
print(len(dataset))

df_2
```

```python
#write in txt dataset
tit, descr, msg, lab = [], [], [], [] 
for item in dataset:
    tit.append(str(item.title))
    descr.append(str(item.description))
    msg.append(str(item.message))
    lab.append(str(item.label))

dataframe = {'Title': tit,
            'Description': descr,
            'Messages': msg,
            'Label': lab
            }

for i in dataset:
    if i.label == "nan" :
        print(i.title)

df_3 = pd.DataFrame(dataframe, columns = ['Title', 'Description', 'Messages', 'Label'])

res = pd.concat([result, df_3])
res = res.sample(frac=1)

print(res['Label'].value_counts())
print(len(res['Label'].value_counts()))

with open("Category_EN_Final.txt", "w") as text_file:
    for i in range(0, len(res)):
        string = '__label__' + str(res.iloc[i, 3]) + ' ' + str(res.iloc[i, 0]) + ' ' + str(res.iloc[i, 1]) + ' ' + str(res.iloc[i, 2]) + '\n'
        text_file.write(string)

with open('Category_EN_Final.txt') as f:
    count = sum(1 for _ in f)

print(count)
        
#create .train and .val
print(int(count*0.85))
print(count - int(count*0.85))
```

```python
newmodel = fasttext.train_supervised(input="Category_EN_Final.train", epoch=50, lr=1.0, wordNgrams=2)
newmodel.save_model("Supervised_EN.bin")
newmodel.test("Category_EN_Final.valid")
```
